```python
# Imporing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading dataset in 'df'

df=pd.read_csv('magic_bricks_cleandata.csv')
```

```python
df
```

```python
# using ordinal encoder

from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df[['Furnishing','District','Status','Transaction','Location','Type']]=oe.fit_transform(df[['Furnishing','District','Status','Transaction','Location','Type']])
```

```python
df
```

```python
# making independent and dependent variables

x=df.drop(['Price','Locality','Unnamed: 0'],axis=1)
y=df.Price
```

```python
# importing train_test_split from model selection
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
# Importing decision tree regressor

from sklearn.tree import DecisionTreeRegressor
```

```python
# putting the algorithm in variable 'dtr'

dtr=DecisionTreeRegressor()
```

```python
# fitting training data

dtr.fit(x_train,y_train)
```

```python
# predicting

y_pred=dtr.predict(x_test)
```

```python
# checking accuracy

from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred).round(2)
```
